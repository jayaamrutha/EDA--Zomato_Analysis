```python
# 7 Infer the relation between rating and Cost of dining
plt.figure(figsize=(12,6))
sns.scatterplot(x="Average Cost for two", y="Aggregate rating", hue='Price range', data=Delhi)

plt.xlabel("Average Cost for two")
plt.ylabel("Aggregate rating")
plt.title('Rating vs Cost of Two');
```

```python
# 8 Get the rating of the restaurants having some common eateries

# Coffee
types = {
    "Coffee" : ["Cafe Coffee Day", "Costa Coffee", "Chaayos"],
    "American": ["Domino's Pizza", "McDonald's", "Burger King", "Subway", "Pizza Hut"]
}

Coffee = Delhi[Delhi['Restaurant Name'].isin(types['Coffee'])]
American = Delhi[Delhi['Restaurant Name'].isin(types['American'])]

Coffee = Coffee[['Restaurant Name','Aggregate rating']].groupby('Restaurant Name').mean().reset_index().sort_values('Aggregate rating',ascending=False)
Coffee


```

```python
# American
American = American[['Restaurant Name','Aggregate rating']].groupby('Restaurant Name').mean().reset_index().sort_values('Aggregate rating',ascending=False)
American
```
